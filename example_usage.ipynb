```python
import pandas as pd

from journal_scrape import JNeurophys
from dbclass import Database
```

### Create a Database object and make sure the _body_ and _metadata_ tables are there.

```python
# initialize the database
db = Database()

# create tables
# this will do nothing if the "body" and "metadata" tables already exist
db.create_tables()

# list the tables
db.list_tables()
```

### Set up the JNeurophys object and scrape the journal

```python
# set up the JNeurophys object
jn = JNeurophys()
```

```python
# scrape the journal
jn.crawl_journal()
```

### Delete the tables
If you need to start from scratch, you can just delete the tables.

```python
sql = """
DROP TABLE metadata, body
"""
with db.conn.cursor() as cursor:
    cursor.execute(sql)
    
db.conn.commit()
```
